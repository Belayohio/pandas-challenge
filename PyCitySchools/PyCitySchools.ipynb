                             # District Summary





```python
import pandas as pd
from pathlib import Path
```

```python
school_data_to_load=Path("Resources/Schools_complete.csv")
student_data_to_load=Path("Resources/Students_complete.csv")
```

```python
school_data=pd.read_csv(school_data_to_load)
school_data
school_data.head()
```

```python
student_data=pd.read_csv(student_data_to_load)
student_data
student_data.head()
```

```python
# lets merge based on the identical column
school_data_complete=pd.merge(student_data,school_data,on ="school_name",how="left")
school_data_complete
school_data_complete.head()
```

```python
# total number of unique school
school_count= len(school_data["school_name"].unique())
school_count
print(f"total number of unique school is {school_count}")
```

```python
#Total students
student_count=student_data["student_name"].count()
print(f' Total number of student for the school district is {student_count}')
```

```python
#to get total budget we need to sum of the budget column.
total_budget=school_data["budget"].sum()
print(f'#the total budget for the district is  {total_budget}')
```

```python
# the Average Math Score
average_math_score=school_data_complete["math_score"].mean()
print(f'# the average math score for the district is {average_math_score}')
```

```python
# Average reading score
average_reading_score=school_data_complete['reading_score'].mean()
print(f' the average reading score for the district is {average_reading_score}')
```

```python
#% of Passing math( the percentage of student who passed math
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(f'the percentage of students who passed math is {passing_math_percentage}%')
```

```python
# % passing reading(the percentage of students who passed reading)
passing_reading_count =school_data_complete[(school_data_complete["reading_score"]>=70)].count()["reading_score"]
passing_reading_percentage =passing_reading_count/float(student_count)*100
print(f' the percenntage of students who passed reading is {passing_reading_percentage}%')
```

```python
# % overall passing(the percentage of student who passed math AND reading)
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
print(f' the percentage of students who passed Math and reading is {overall_passing_rate}%')
```

```python
district_summary =pd.DataFrame([{"Total Schools":school_count,"Total Students":student_count,"Total Budget":total_budget,"Average_Math_score":average_math_score,
                    "Average_reading_score":average_reading_score,"% Passing math":passing_math_percentage,
                   "% Passing Reading":passing_reading_percentage,
                   "% Overall Passing":overall_passing_rate}])

                   
# Formatting
district_summary["Total Students"]= district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary
```

# Scores by School Spending
#Create a table that breaks down school performance based on average spending ranges (per student).



```python
school_spending_df=per_school_summary.copy()
```

```python
spending_bins =  [0, 585, 630, 645, 680]
```

```python
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_df["Spending Ranges (Per Student)"]=pd.cut(school_spending_df["Per Student Budget"].str[1:].astype(float),bins=spending_bins,labels=labels)
school_spending_df["Spending Ranges (Per Student)"] = school_spending_df["Spending Ranges (Per Student)"].astype(str)
school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
```

```python
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
```

```python
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
```

```python
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
```

```python
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
spending_summary=pd.DataFrame({"Average Math Score":spending_math_scores,"Average Reading Score":spending_reading_scores,"% Passing Math":spending_passing_math
    ,"% Passing Reading":spending_passing_reading,"% Overall Passing": overall_passing_spending})
spending_summary
```

# Scores by School Size
#Use the following code to create three bins with reasonable cutoff values to group school size.




```python
size_bins = [0, 1000, 2000, 5000]
```

```python
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_size_df = per_school_summary.copy()
```

```python
school_size_df["School Size"] =pd.cut(school_size_df["Total Students"],bins=size_bins,labels=labels,include_lowest=True)
school_size_df["School Size"]=school_size_df["School Size"].astype(str)
school_size_df
```

```python
size_math_scores = school_size_df.groupby(["School Size"])["Average Math Score"].mean()
```

```python
size_reading_scores = school_size_df.groupby(["School Size"])["Average Reading Score"].mean()
```

```python
size_passing_math = school_size_df.groupby(["School Size"])["% Passing Math"].mean()
```

```python
size_passing_reading = school_size_df.groupby(["School Size"])["% Passing Reading"].mean()
```

```python
size_overall_passing = school_size_df.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
size_summary=pd.DataFrame({"Average Math Score":size_math_scores,"Average Reading Score":size_reading_scores,
                          "% Passing Math":size_passing_math,"% Passing Reading":size_passing_reading,"% Overall Passing":size_overall_passing})
size_summary
```
